```python
import speech_recognition as sr

# Initialize recognizer (this handles speech processing)
recognizer = sr.Recognizer()

# Load audio file (make sure this file exists in the same folder)
audio_file = "audio.wav"   # keep your audio in same folder

# Open the audio file as a source
with sr.AudioFile(audio_file) as source:
    print("Listening...")  # Indicates that audio is being processed
    
    # Record the entire audio file
    audio = recognizer.record(source)

# -------------------- Speech to Text --------------------

# Try converting the audio into text using Google Speech API
try:
    text = recognizer.recognize_google(audio)
    
    # Print the recognized text
    print("Transcription:", text)

# If speech is unclear or not recognized
except sr.UnknownValueError:
    print("Could not understand audio")

# If there is an issue with the API (e.g., no internet)
except sr.RequestError:
    print("API unavailable")
```

```python
import speech_recognition as sr

# Create a recognizer object (used to process audio)
recognizer = sr.Recognizer()

# Open the audio file (make sure "audio.wav" is in the same folder)
with sr.AudioFile("audio.wav") as source:
    print("Processing audio...")  # Indicates that audio is being read
    
    # Record the complete audio from the file
    audio = recognizer.record(source)

# -------------------- Speech Recognition --------------------

try:
    # Convert speech in audio to text using Google Speech API
    text = recognizer.recognize_google(audio)
    
    # Print the recognized speech
    print("You said:", text)

# If any error occurs (could be unclear audio or API issue)
except:
    print("Error recognizing speech")
```
